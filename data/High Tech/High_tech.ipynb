```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

amadeus = pd.read_csv('ams.csv') 
indra = pd.read_csv('idr.csv')
```

```python
# Paso 2: Concatenar los diferentes DFs para tener un consolidado de High Tech

high_tech = pd.concat([amadeus, indra], axis=0)

high_tech.reset_index(drop=True)
```

```python
# Paso 3: Exportar el DF a csv

high_tech.to_csv('ht.csv')
```
