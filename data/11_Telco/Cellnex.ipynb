```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

cellnex = pd.read_csv('CLNX.MC.csv')

cellnex
```

```python
cellnex['id'] = 13
cellnex['ticker'] = 'CLNX'
cellnex['empresa'] = 'Cellnex'

cellnex
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


cellnex.columns = cellnex.columns.str.lower()

cellnex.columns
```

```python
# Paso 4: Ordenar columnas del DF

cellnex = cellnex[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

cellnex
```

```python
# Paso 5: Pasar a csv el DF

cellnex.to_csv('clnx.csv',index=False)
```
