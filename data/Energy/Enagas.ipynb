```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

enagas = pd.read_csv('ENG.MC.csv')

enagas
```

```python
enagas['id'] = 14
enagas['ticker'] = 'ENG'
enagas['empresa'] = 'Enagas'

enagas
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


enagas.columns = enagas.columns.str.lower()

enagas.columns
```

```python
# Paso 4: Ordenar columnas del DF

enagas = enagas[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

enagas
```

```python
# Paso 5: Pasar a csv el DF

enagas.to_csv('eng.csv',index=False)
```
