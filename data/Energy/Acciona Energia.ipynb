```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

acciona_ene = pd.read_csv('ANE.MC.csv')

acciona_ene
```

```python
acciona_ene['id'] = 2
acciona_ene['ticker'] = 'ANE'
acciona_ene['empresa'] = 'Acciona Energia'

acciona_ene
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


acciona_ene.columns = acciona_ene.columns.str.lower()

acciona_ene.columns
```

```python
# Paso 4: Ordenar columnas del DF

acciona_ene = acciona_ene[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

acciona_ene
```

```python
# Paso 5: Pasar a csv el DF

acciona_ene.to_csv('ane.csv',index=False)
```
