```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

fluidra = pd.read_csv('FDR.MC.csv')

fluidra
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

fluidra['id'] = 17
fluidra['ticker'] = 'FDR'
fluidra['empresa'] = 'Fluidra'

fluidra
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


fluidra.columns = fluidra.columns.str.lower()

fluidra.columns
```

```python
# Paso 4: Ordenar columnas del DF

fluidra = fluidra[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

fluidra
```

```python
# Paso 5: Pasar a csv el DF

fluidra.to_csv('fdr.csv',index=False)
```
