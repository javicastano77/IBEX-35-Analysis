```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

repsol = pd.read_csv('REP.MC.csv')

repsol
```

```python
repsol['id'] = 30
repsol['ticker'] = 'REP'
repsol['empresa'] = 'Repsol'

repsol
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


repsol.columns = repsol.columns.str.lower()

repsol.columns
```

```python
# Paso 4: Ordenar columnas del DF

repsol = repsol[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

repsol
```

```python
# Paso 5: Pasar a csv el DF

repsol.to_csv('rep.csv',index=False)
```
