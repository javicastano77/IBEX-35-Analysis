```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

naturgy = pd.read_csv('NTGY.MC.csv')

naturgy
```

```python
naturgy['id'] = 28
naturgy['ticker'] = 'NTGY'
naturgy['empresa'] = 'Naturgy'

naturgy
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


naturgy.columns = naturgy.columns.str.lower()

naturgy.columns
```

```python
# Paso 4: Ordenar columnas del DF

naturgy = naturgy[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

naturgy
```

```python
# Paso 5: Pasar a csv el DF

naturgy.to_csv('ntgy.csv',index=False)
```
