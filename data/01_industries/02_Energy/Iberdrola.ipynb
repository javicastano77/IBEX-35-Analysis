```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

iberdrola = pd.read_csv('IBE.MC.csv')

iberdrola
```

```python
iberdrola['id'] = 20
iberdrola['ticker'] = 'IBE'
iberdrola['empresa'] = 'Iberdrola'

iberdrola
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


iberdrola.columns = iberdrola.columns.str.lower()

iberdrola.columns
```

```python
# Paso 4: Ordenar columnas del DF

iberdrola = iberdrola[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

iberdrola
```

```python
# Paso 5: Pasar a csv el DF

iberdrola.to_csv('ibe.csv',index=False)
```
