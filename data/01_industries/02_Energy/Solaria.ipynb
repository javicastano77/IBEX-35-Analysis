```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

solaria = pd.read_csv('SLR.MC.csv')

solaria
```

```python
solaria['id'] = 33
solaria['ticker'] = 'SLR'
solaria['empresa'] = 'Solaria'

solaria
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


solaria.columns = solaria.columns.str.lower()

solaria.columns
```

```python
# Paso 4: Ordenar columnas del DF

solaria = solaria[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

solaria
```

```python
# Paso 5: Pasar a csv el DF

solaria.to_csv('slr.csv',index=False)
```
