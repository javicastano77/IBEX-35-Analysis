```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

colonial = pd.read_csv('COL.MC.csv')

colonial
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

colonial['id'] = 23
colonial['ticker'] = 'COL'
colonial['empresa'] = 'Colonial Inmobiliaria'

colonial
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


colonial.columns = colonial.columns.str.lower()

colonial.columns
```

```python
# Paso 4: Ordenar columnas del DF

colonial = colonial[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

colonial
```

```python
# Paso 5: Pasar a csv el DF

colonial.to_csv('col.csv',index=False)
```
