```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

indra = pd.read_csv('IDR.MC.csv')

indra
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

indra['id'] = 22
indra['ticker'] = 'IDR'
indra['empresa'] = 'Indra'

indra
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


indra.columns = indra.columns.str.lower()

indra.columns
```

```python
# Paso 4: Ordenar columnas del DF

indra = indra[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

indra
```

```python
# Paso 5: Pasar a csv el DF

indra.to_csv('idr.csv',index=False)
```
