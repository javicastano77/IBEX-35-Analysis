```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

telefonica = pd.read_csv('TEF.MC.csv')

telefonica
```

```python
telefonica['id'] = 34
telefonica['ticker'] = 'TEF'
telefonica['empresa'] = 'Telefonica'

telefonica
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


telefonica.columns = telefonica.columns.str.lower()

telefonica.columns
```

```python
# Paso 4: Ordenar columnas del DF

telefonica = telefonica[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

telefonica
```

```python
# Paso 5: Pasar a csv el DF

telefonica.to_csv('tef.csv',index=False)
```
