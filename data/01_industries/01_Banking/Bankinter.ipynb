```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

bankinter = pd.read_csv('BKT.MC.csv')

bankinter
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

bankinter['id'] = 10
bankinter['ticker'] = 'BKT'
bankinter['empresa'] = 'Bankinter'

bankinter
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


bankinter.columns = bankinter.columns.str.lower()

bankinter.columns
```

```python
# Paso 4: Ordenar columnas del DF

bankinter = bankinter[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

bankinter
```

```python
# Paso 5: Pasar a csv el DF

bankinter.to_csv('bkt.csv',index=False)
```
