```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

bankinter = pd.read_csv('bkt.csv')
bbva = pd.read_csv('bbva.csv')
caixabank = pd.read_csv('cabk.csv')
sabadell = pd.read_csv('sab.csv')
santander = pd.read_csv('san.csv')
unicaja = pd.read_csv('uni.csv')

```

```python
# Paso 2: Concatenar los diferentes DFs para tener un consolidado de Banking

banking = pd.concat([bankinter, bbva,caixabank,sabadell,santander,unicaja], axis=0)

banking.reset_index(drop=True)
```

```python
# Paso 3: Exportar el DF a csv

banking.to_csv('banking.csv')
```
