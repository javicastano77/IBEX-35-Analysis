```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

banking = pd.read_csv('banking.csv')
energy = pd.read_csv('energy.csv')
high_tech = pd.read_csv('ht.csv')
industrial_services = pd.read_csv('industrial.csv')
insurance = pd.read_csv('map.csv')
investment = pd.read_csv('investment.csv')
logistics = pd.read_csv('log.csv')
pharma = pd.read_csv('pharma.csv')
resources = pd.read_csv('resources.csv')
retail = pd.read_csv('itx.csv')
telco = pd.read_csv('telco.csv')
travel = pd.read_csv('travel.csv')
```

```python
# Paso 2: Concatenar los diferentes consolidados para tener un DF con todas las empresas del Ibex35

ibex35 = pd.concat([banking, energy,high_tech,industrial_services,insurance,investment,logistics,pharma,
                    resources,retail,telco,travel], axis=0)

ibex35 = ibex35.reset_index(drop=True)
ibex35 = ibex35.drop(['Unnamed: 0'],axis=1)
```

```python
# Paso 3: Ordenar el DF por la columna 'id'

ibex35 = ibex35.sort_values(by='id',ascending=True)

ibex35 = ibex35.reset_index(drop=True)

ibex35
```

```python
# Paso 3: Ordenar el DF por la columna 'id'

ibex35.to_csv('ibex35.csv',index=False)
```
