```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

colonial = pd.read_csv('col.csv')
merlin = pd.read_csv('mrl.csv')
```

```python
# Paso 2: Concatenar los diferentes DFs para tener un consolidado de Banking

investment = pd.concat([colonial, merlin], axis=0)

investment.reset_index(drop=True)
```

```python
# Paso 3: Exportar el DF a csv

investment.to_csv('investment.csv')
```
