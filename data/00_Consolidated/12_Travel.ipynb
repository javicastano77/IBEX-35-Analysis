```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

aena = pd.read_csv('aena.csv')
melia = pd.read_csv('mel.csv')
```

```python
# Paso 2: Concatenar los diferentes DFs para tener un consolidado de Banking

travel = pd.concat([aena, melia], axis=0)

travel.reset_index(drop=True)
```

```python
# Paso 3: Exportar el DF a csv

travel.to_csv('travel.csv')
```
