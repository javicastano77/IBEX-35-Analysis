```python
# Paso 0: Importar libreria Pandas para cargar DF

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

acciona_ene = pd.read_csv('ane.csv')
enagas = pd.read_csv('eng.csv')
endesa =  pd.read_csv('ele.csv')
iberdrola = pd.read_csv('ibe.csv')
naturgy = pd.read_csv('ntgy.csv')
ree = pd.read_csv('red.csv')
repsol = pd.read_csv('rep.csv')
solaria = pd.read_csv('slr.csv')

```

```python
# Paso 2: Concatenar los diferentes DFs para tener un consolidado de Energy

energy = pd.concat([acciona_ene, enagas,endesa,iberdrola,naturgy,ree,repsol,solaria], axis=0)

energy.reset_index(drop=True)
```

```python
# Paso 3: Exportar el DF a csv

energy.to_csv('energy.csv')
```
