```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

mapfre = pd.read_csv('MAP.MC.csv')

mapfre
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

mapfre['id'] = 25
mapfre['ticker'] = 'MAP'
mapfre['empresa'] = 'mapfre'

mapfre
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


mapfre.columns = mapfre.columns.str.lower()

mapfre.columns
```

```python
# Paso 4: Ordenar columnas del DF

mapfre = mapfre[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

mapfre
```

```python
# Paso 5: Pasar a csv el DF

mapfre.to_csv('map.csv',index=False)
```
