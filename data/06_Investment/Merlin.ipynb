```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

merlin = pd.read_csv('MRL.MC.csv')

merlin
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

merlin['id'] = 27
merlin['ticker'] = 'MRL'
merlin['empresa'] = 'Merlin Properties'

merlin
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


merlin.columns = merlin.columns.str.lower()

merlin.columns
```

```python
# Paso 4: Ordenar columnas del DF

merlin = merlin[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

merlin
```

```python
# Paso 5: Pasar a csv el DF

merlin.to_csv('mrl.csv',index=False)
```
