```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

logista = pd.read_csv('LOG.MC.csv')

logista
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

logista['id'] = 24
logista['ticker'] = 'LOG'
logista['empresa'] = 'Logista'

logista
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


logista.columns = logista.columns.str.lower()

logista.columns
```

```python
# Paso 4: Ordenar columnas del DF

logista = logista[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

logista
```

```python
# Paso 5: Pasar a csv el DF

logista.to_csv('log.csv',index=False)
```
