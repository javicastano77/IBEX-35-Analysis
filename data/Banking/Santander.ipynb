```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

santander = pd.read_csv('SAN.MC.csv')

santander
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

santander['id'] = 9
santander['ticker'] = 'SAN'
santander['empresa'] = 'Santander'

santander
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


santander.columns = santander.columns.str.lower()

santander.columns
```

```python
# Paso 4: Ordenar columnas del DF

santander = santander[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

santander
```

```python
# Paso 5: Pasar a csv el DF

santander.to_csv('san.csv',index=False)
```
