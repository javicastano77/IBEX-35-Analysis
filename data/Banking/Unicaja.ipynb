```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

unicaja = pd.read_csv('UNI.MC.csv')

unicaja
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

unicaja['id'] = 35
unicaja['ticker'] = 'UNI'
unicaja['empresa'] = 'Unicaja'

unicaja
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


unicaja.columns = unicaja.columns.str.lower()

unicaja.columns
```

```python
# Paso 4: Ordenar columnas del DF

unicaja = unicaja[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

unicaja
```

```python
# Paso 5: Pasar a csv el DF

unicaja.to_csv('uni.csv',index=False)
```
