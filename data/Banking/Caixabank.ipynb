```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

caixabank = pd.read_csv('CABK.MC.csv')

caixabank
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

caixabank['id'] = 12
caixabank['ticker'] = 'CABK'
caixabank['empresa'] = 'Caixabank'

caixabank
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


caixabank.columns = caixabank.columns.str.lower()

caixabank.columns
```

```python
# Paso 4: Ordenar columnas del DF

caixabank = caixabank[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

caixabank
```

```python
# Paso 5: Pasar a csv el DF

caixabank.to_csv('cabk.csv',index=False)
```
