```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

sabadell = pd.read_csv('SAB.MC.csv')

sabadell
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

sabadell['id'] = 8
sabadell['ticker'] = 'SAB'
sabadell['empresa'] = 'Sabadell'

sabadell
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


sabadell.columns = sabadell.columns.str.lower()

sabadell.columns
```

```python
# Paso 4: Ordenar columnas del DF

sabadell = sabadell[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

sabadell
```

```python
# Paso 5: Pasar a csv el DF

sabadell.to_csv('sab.csv',index=False)
```
