```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

bbva = pd.read_csv('BBVA.MC.csv')

bbva
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

bbva['id'] = 11
bbva['ticker'] = 'BBVA'
bbva['empresa'] = 'BBVA'

bbva
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


bbva.columns = bbva.columns.str.lower()

bbva.columns
```

```python
# Paso 4: Ordenar columnas del DF

bbva = bbva[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

bbva
```

```python
# Paso 5: Pasar a csv el DF

bbva.to_csv('bbva.csv',index=False)
```
