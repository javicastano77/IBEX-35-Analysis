```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

rovi = pd.read_csv('ROVI.MC.csv')

rovi
```

```python
rovi['id'] = 31
rovi['ticker'] = 'ROVI'
rovi['empresa'] = 'Laboratorios Rovi'

rovi
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


rovi.columns = rovi.columns.str.lower()

rovi.columns
```

```python
# Paso 4: Ordenar columnas del DF

rovi = rovi[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

rovi
```

```python
# Paso 5: Pasar a csv el DF

rovi.to_csv('rovi.csv',index=False)
```
