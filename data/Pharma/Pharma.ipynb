```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

grifols = pd.read_csv('grf.csv')
rovi = pd.read_csv('rovi.csv')
```

```python
# Paso 2: Concatenar los diferentes DFs para tener un consolidado de Banking

pharma = pd.concat([grifols, rovi], axis=0)

pharma.reset_index(drop=True)
```

```python
# Paso 3: Exportar el DF a csv

pharma.to_csv('pharma.csv')
```
