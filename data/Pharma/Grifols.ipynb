```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

grifols = pd.read_csv('GRF.MC.csv')

grifols
```

```python
grifols['id'] = 18
grifols['ticker'] = 'GRF'
grifols['empresa'] = 'Grifols'

grifols
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


grifols.columns = grifols.columns.str.lower()

grifols.columns
```

```python
# Paso 4: Ordenar columnas del DF

grifols = grifols[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

grifols
```

```python
# Paso 5: Pasar a csv el DF

grifols.to_csv('grf.csv',index=False)
```
