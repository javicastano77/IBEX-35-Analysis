```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

acs = pd.read_csv('ACS.MC.csv')

acs
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

acs['id'] = 4
acs['ticker'] = 'ACS'
acs['empresa'] = 'Grupo ACS'

acs
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


acs.columns = acs.columns.str.lower()

acs.columns
```

```python
# Paso 4: Ordenar columnas del DF

acs = acs[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

acs
```

```python
# Paso 5: Pasar a csv el DF

acs.to_csv('acs.csv',index=False)
```
