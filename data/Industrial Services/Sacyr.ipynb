```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

sacyr = pd.read_csv('SCYR.MC.csv')

sacyr
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

sacyr['id'] = 32
sacyr['ticker'] = 'SCYR'
sacyr['empresa'] = 'Sacyr'

sacyr
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


sacyr.columns = sacyr.columns.str.lower()

sacyr.columns
```

```python
# Paso 4: Ordenar columnas del DF

sacyr = sacyr[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

sacyr
```

```python
# Paso 5: Pasar a csv el DF

sacyr.to_csv('scyr.csv',index=False)
```
