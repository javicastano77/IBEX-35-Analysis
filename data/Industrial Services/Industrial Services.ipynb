```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

acciona = pd.read_csv('ana.csv')
ferrovial = pd.read_csv('fer.csv')
fluidra = pd.read_csv('fdr.csv')
grupo_acs = pd.read_csv('acs.csv')
sacyr = pd.read_csv('scyr.csv')
```

```python
# Paso 2: Concatenar los diferentes DFs para tener un consolidado de Banking

industrial_services = pd.concat([acciona, ferrovial,fluidra,grupo_acs,sacyr], axis=0)

industrial_services.reset_index(drop=True)
```

```python
# Paso 3: Exportar el DF a csv

industrial_services.to_csv('industrial.csv')
```
