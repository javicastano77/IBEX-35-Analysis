```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

ferrovial = pd.read_csv('FER.MC.csv')

ferrovial
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

ferrovial['id'] = 16
ferrovial['ticker'] = 'FER'
ferrovial['empresa'] = 'Ferrovial'

ferrovial
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


ferrovial.columns = ferrovial.columns.str.lower()

ferrovial.columns
```

```python
# Paso 4: Ordenar columnas del DF

ferrovial = ferrovial[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

ferrovial
```

```python
# Paso 5: Pasar a csv el DF

ferrovial.to_csv('fer.csv',index=False)
```
