```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

ana = pd.read_csv('AJ3.BE.csv')

ana
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

ana['id'] = 1
ana['ticker'] = 'ANA'
ana['empresa'] = 'Acciona'

ana
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


ana.columns = ana.columns.str.lower()

ana.columns
```

```python
# Paso 4: Ordenar columnas del DF

ana = ana[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

ana
```

```python
# Paso 5: Pasar a csv el DF

ana.to_csv('ana.csv',index=False)
```
