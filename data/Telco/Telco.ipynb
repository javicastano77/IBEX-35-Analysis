```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

cellnex = pd.read_csv('clnx.csv')
telefonica = pd.read_csv('tef.csv')
```

```python
# Paso 2: Concatenar los diferentes DFs para tener un consolidado de Banking

telco = pd.concat([cellnex, telefonica], axis=0)

telco.reset_index(drop=True)
```

```python
# Paso 3: Exportar el DF a csv

telco.to_csv('telco.csv')
```
