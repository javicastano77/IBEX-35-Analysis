```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

amadeus = pd.read_csv('AMS.MC.csv')

amadeus
```

```python
# Paso 2: Insertar nuevas columnas en el DF, 'id','ticker','empresa'

amadeus['id'] = 6
amadeus['ticker'] = 'AMS'
amadeus['empresa'] = 'Amadeus'

amadeus
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


amadeus.columns = amadeus.columns.str.lower()

amadeus.columns
```

```python
# Paso 4: Ordenar columnas del DF

amadeus = amadeus[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

amadeus
```

```python
# Paso 5: Pasar a csv el DF

amadeus.to_csv('ams.csv',index=False)
```
