```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

aena = pd.read_csv('AENA.MC.csv')

aena
```

```python
aena['id'] = 5
aena['ticker'] = 'AENA'
aena['empresa'] = 'Aena'

aena
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


aena.columns = aena.columns.str.lower()

aena.columns
```

```python
# Paso 4: Ordenar columnas del DF

aena = aena[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

aena
```

```python
# Paso 5: Pasar a csv el DF

aena.to_csv('aena.csv',index=False)
```
