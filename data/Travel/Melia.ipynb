```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

melia = pd.read_csv('MEL.DU.csv')

melia
```

```python
melia['id'] = 26
melia['ticker'] = 'MEL'
melia['empresa'] = 'Melia'

melia
```

```python
# Paso 3: Modificar tÃ­tulos de columnas con funciÃ³n lower.


melia.columns = melia.columns.str.lower()

melia.columns
```

```python
# Paso 4: Ordenar columnas del DF

melia = melia[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

melia
```

```python
# Paso 5: Pasar a csv el DF

melia.to_csv('mel.csv',index=False)
```
