```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

acerinox = pd.read_csv('ACX.MC.csv')

acerinox
```

```python
acerinox['id'] = 3
acerinox['ticker'] = 'ACX'
acerinox['empresa'] = 'Acerinox'

acerinox
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


acerinox.columns = acerinox.columns.str.lower()

acerinox.columns
```

```python
# Paso 4: Ordenar columnas del DF

acerinox = acerinox[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

acerinox
```

```python
# Paso 5: Pasar a csv el DF

acerinox.to_csv('acx.csv',index=False)
```
