```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DFs con csv generados

acerinox = pd.read_csv('acx.csv')
arcelor = pd.read_csv('mts.csv')
```

```python
# Paso 2: Concatenar los diferentes DFs para tener un consolidado de Banking

resources = pd.concat([acerinox, arcelor], axis=0)

resources.reset_index(drop=True)
```

```python
# Paso 3: Exportar el DF a csv

resources.to_csv('resources.csv')
```
