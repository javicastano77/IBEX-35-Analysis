```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 1: Cargar DF con la info descargada

arcelor = pd.read_csv('MTS.MC.csv')

arcelor
```

```python
arcelor['id'] = 7
arcelor['ticker'] = 'MTS'
arcelor['empresa'] = 'Arcelor'

arcelor
```

```python
# Paso 3: Modificar títulos de columnas con función lower.


arcelor.columns = arcelor.columns.str.lower()

arcelor.columns
```

```python
# Paso 4: Ordenar columnas del DF

arcelor = arcelor[['id','ticker','empresa','date','open','high','low','close','adj close','volume']]

arcelor
```

```python
# Paso 5: Pasar a csv el DF

arcelor.to_csv('mts.csv',index=False)
```
