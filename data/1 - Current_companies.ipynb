```python
# Paso 0: Importar libreria Pandas para cargar DF

import pandas as pd
```

```python
# Paso 0: Modificar el display para ver todas las filas

pd.options.display.max_rows=None
```

```python
# Paso 1: Realizar scrapping de Tabla de Wikipedia sobre las empresas actuales del IBEX 35

url = 'https://es.wikipedia.org/wiki/IBEX_35'

df2 = pd.read_html(url)[1]
```

```python
# Paso 2: Crear nueva columna para chequear que son 35 valores

df2['id'] = df2.index+1
```

```python
df2
```

```python
# Paso 3: Ubicar el 'id' de la empresa como la primera columna del DF

df2.insert(0,'id', df2.pop('id'))

df2
```

```python
# Paso 5: Modificar títulos de columnas con función lower.


df2.columns = df2.columns.str.lower()

df2.columns
```

```python
# Paso 6: Generar nueva columna de sector sustituyendo a 'sector[28]'.

df2 ['sector'] = df2['sector[28]​']

df2 = df2.drop(['sector[28]​'],axis=1)

df2.insert(3,'sector', df2.pop('sector'))

df2
```

```python
# Paso 7: Analizar valores únicos de variable 'sector'

df2.sector.value_counts()
```

```python
# Paso 8: Renombrar valores únicos de variables 'sector' para mejorar consistencia

# Crear diccionario para la nueva codificación

codi = {'Bancos y cajas de ahorro': 'Banking','Electricidad y gas': 'Energy','Construcción':'Industrial Services',
          'Energías renovables': 'Energy', 'Mineral, metales y transformación': 'Resources', 'Transporte y distribución':'Logistic',
          'Electrónica y software':'High Tech','Telecomunicaciones y otros':'Telco','Productos farmacéuticos y biotecnología':'Pharma',
          'SOCIMI':'Investment','Ingeniería y otros':'Industrial Services','Textil, vestido y calzado':'Retail',
          'Transporte y Distribución':'Logistic','Seguros':'Insurance','Ocio, turismo y hostelería':'Travel','Petróleo':'Energy'}

# Reemplazar los valores

df2['sector'] = df2['sector'].replace(codi)
```

```python
# Paso 9: Renombrar algunos valores únicos de variable 'empresa' para evitar errores de codificación

# Crear diccionario para la nueva codificación

companies = {'Amadeus IT Group':'Amadeus','ArcelorMittal':'Arcelor','Cellnex Telecom':'Cellnex','Enagás':'Enagas','Indra Sistemas':'Indra',
            'Meliá Hotels International':'Melia','Red Eléctrica Corporación':'REE','Solaria Energía y Medio Ambiente':'Solaria',
             'Telefónica':'Telefonica','Unicaja Banco':'Unicaja'}

# Reemplazar los valores

df2['empresa'] = df2['empresa'].replace(companies)


```

```python
df2
```

```python
# Paso 10: Eliminar columnas del DF que no aportan

df2 = df2[['id','ticker','empresa','sector','sede','entrada']]

df2
```
