```python
# Paso 0: Importar libreria Pandas para

import pandas as pd
```

```python
# Paso 1: Generar DF a través del csv importado

df = pd.read_csv('ibex35.csv')

df
```

```python
# Paso 2: Chequear los nombres de las empresas cargadas en el Ibex35

df.empresa.unique()
```

```python
# Paso 3: Crear un diccionario para clasificar las empresas en sectores para nueva columna

category = {'Acciona':'construction','Grupo ACS':'construction','Ferrovial':'construction','Sacyr':'construction','Fluidra':'construction',
             'Acciona Energia':'energy','Enagas':'energy','Endesa':'energy','Iberdrola':'energy','Naturgy':'energy','REE':'energy','Solaria':'energy',
             'Sabadell':'banking','Santander':'banking','Bankinter':'banking','BBVA':'banking','Caixabank':'banking','Unicaja':'banking',
             'Acerinox':'resources','Arcelor':'resources','Repsol':'resources',
             'Cellnex':'telco','Telefonica':'telco',
             'Amadeus':'high tech','Indra':'high tech',
             'Aena':'transport','IAG':'transport','Logista':'transport',
             'Grifols':'pharma','Laboratorios Rovi':'pharma',
             'Melia':'travel','mapfre':'insurance',
             'Merlin Properties':'real estate','Colonial Inmobiliaria':'real estate',
             'Inditex':'retail'}

df['sector'] = df['empresa'].map(category)

df
```

```python
# Paso 4: Comprobar formato del DF

df.info()
```

```python
# Paso 5: Modificar el formato de fecha a datetime

df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")

df = df.sort_values(by=['id','date'],ascending= True)

df
```

```python
# Paso 6: Sustituir '.' por ',' para poder trabajar en Power BI

df = df.applymap(lambda x: x if not isinstance(x, (int, float)) else str(x).replace('.', ','))

df
```

```python
# Paso 7: Exportar DF a csv

df.to_csv('ibex_coma.csv',index=False)
```
