```python
# Paso 0: Importar libreria Pandas

import pandas as pd
```

```python
# Paso 2: Importar resultados del modelo para cada empresa


acciona = pd.read_csv('acciona.csv')
acciona_ene = pd.read_csv('acciona_ene.csv')
acerinox = pd.read_csv('acerinox.csv')
acs = pd.read_csv('acs.csv')
aena = pd.read_csv('aena.csv')
amadeus = pd.read_csv('amadeus.csv')
arcelor = pd.read_csv('arcelor.csv')
bankinter = pd.read_csv('bankinter.csv')
bbva = pd.read_csv('bbva.csv')
caixabank = pd.read_csv('caixabank.csv')
cellnex = pd.read_csv('cellnex.csv')
colonial = pd.read_csv('colonial.csv')
enagas = pd.read_csv('enagas.csv')
endesa = pd.read_csv('endesa.csv')
ferrovial = pd.read_csv('ferrovial.csv')
fluidra = pd.read_csv('fluidra.csv') 
grifols = pd.read_csv('grifols.csv')
iag = pd.read_csv('iag.csv') 
iberdrola = pd.read_csv('iberdrola.csv')
inditex = pd.read_csv('inditex.csv')
indra = pd.read_csv('indra.csv')
logista = pd.read_csv('logista.csv')
mapfre = pd.read_csv('mapfre.csv')
melia = pd.read_csv('melia.csv') 
merlin = pd.read_csv('merlin.csv')
naturgy = pd.read_csv('naturgy.csv')
ree = pd.read_csv('ree.csv')
repsol = pd.read_csv('repsol.csv')
rovi = pd.read_csv('rovi.csv')
sabadell = pd.read_csv('sabadell.csv')
sacyr = pd.read_csv('sacyr.csv') 
santander = pd.read_csv('santander.csv')
solaria = pd.read_csv('solaria.csv')
telefonica = pd.read_csv('telefonica.csv')
unicaja = pd.read_csv('unicaja.csv')
```

```python
# Paso 2: Realizar concatenación para todos los DFs importados

list_of_dfs = [acciona, acciona_ene, acerinox, acs, aena, amadeus, arcelor, bankinter, bbva, caixabank, cellnex, colonial, enagas, endesa, 
               ferrovial, fluidra, grifols, iag, iberdrola, inditex, indra, logista, mapfre, melia, merlin, naturgy,ree,repsol,
               rovi, sabadell, sacyr, santander, solaria, telefonica, unicaja]


simulator = pd.concat(list_of_dfs, ignore_index=True)

simulator
```

```python
# Paso 3: Chequear el tipo del dato del DF resultante

simulator.info()
```

```python
# Paso 4: Sustituir los '.' por ',' para poder trabajar posteriormente en Power BI

columns_to_convert = simulator.columns.difference(['name', 'date'])  # Excluir 'name' y 'date'
simulator[columns_to_convert] = simulator[columns_to_convert].astype(str).apply(lambda x: x.str.replace('.', ','))
```

```python
# Paso 5: Modificar el formato de fecha de string a dattime

simulator['date'] = pd.to_datetime(simulator['date'], format="%Y-%m-%d")

simulator.info()
```

```python
# Paso 6: Realizar último check al DF

simulator
```

```python
# Paso 7: Exportar DF a formato csv para visuzalización en Power BI

simulator.to_csv('simulador.csv')
```
