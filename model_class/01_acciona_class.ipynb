```python
import pandas as pd
```

```python
acciona_class = pd.read_csv('acciona_class.csv')

acciona_class
```

```python
X= acciona_class[['d-4','d-3','d-2','d-1','real']]
y = acciona_class.cate
```

```python
from sklearn.model_selection import train_test_split as tts   


X_train, X_test, y_train, y_test = tts(X, y,
                                       train_size=0.8,
                                       test_size=0.2,
                                       random_state=42
                                      )

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
%pip install lazypredict
```

```python
from lazypredict.Supervised import LazyClassifier

# Initialize Lazy Classifier
classy = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
# Fit and score all models
models, predictions = classy.fit(X_train, X_test, y_train, y_test)
print(models)

```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from sklearn.metrics import classification_report
```

```python
qda_model = QDA()
```

```python
qda_model.fit(X_train, y_train)
```

```python
y_pred = qda_model.predict(X_test)
```

```python
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred))
```
